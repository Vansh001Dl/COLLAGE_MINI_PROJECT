```python
df.replace(["", "Unknown", "N/A", "Missing", "Not Available", "--"], np.nan, inplace=True)
print(df)
```

```python
df.isnull().sum()
```

```python
df.dropna(how="all", inplace=True)
```

```python
print(df)
```

```python
df.dropna(thresh=len(df)*0.5, axis=1, inplace=True)
```

```python
df=df.round(1)
```

```python













df.fillna(df.mean(numeric_only=True), inplace=True)
df.fillna(df.mode().iloc[0], inplace=True)
print(df)
```

```python
# df.fillna(df.mode().iloc[0], inplace=True)
# print(df)
```

```python
df.replace(["", "Unknown", "N/A", "Missing", "Not Available", "--"], np.nan, inplace=True)
```

```python
cols = ["Lucknow", "Delhi", "Kanpur", "Mumbai", "Noida"]
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
```

```python
df.interpolate(method='linear', inplace=True)
```

```python
df[cols] = df[cols].round(2)
```

```python
lucknow = df["Lucknow"].to_numpy()
delhi   = df["Delhi"].to_numpy()
kanpur  = df["Kanpur"].to_numpy()
mumbai  = df["Mumbai"].to_numpy()
noida   = df["Noida"].to_numpy()
print(lucknow)
print(delhi)
print(kanpur)
print(mumbai)
print(noida)
```

```python
months = df["Month"].astype(str).to_numpy()
print(months)
```

```python
# sum of temperature
print("Lucknow:", np.sum(lucknow))
print("Delhi:",   np.sum(delhi))
print("Kanpur:",  np.sum(kanpur))
print("Mumbai:",  np.sum(mumbai))
print("Noida:",   np.sum(noida))
```

```python
# Mean of temperature city wise
print("Lucknow:", np.mean(lucknow))
print("Delhi:",   np.mean(delhi))
print("Kanpur:",  np.mean(kanpur))
print("Mumbai:",  np.mean(mumbai))
print("Noida:",   np.mean(noida))
```

```python
# Max value of temperature
print("Lucknow:", np.max(lucknow))
print("Delhi:",   np.max(delhi))
print("Kanpur:",  np.max(kanpur))
print("Mumbai:",  np.max(mumbai))
print("Noida:",   np.max(noida))
```

```python
# Median of temperature
print("Lucknow:", np.median(lucknow))
print("Delhi:",   np.median(delhi))
print("Kanpur:",  np.median(kanpur))
print("Mumbai:",  np.median(mumbai))
print("Noida:",   np.median(noida))
```

```python
# Standard Deviation
print("Lucknow:", np.std(lucknow))
print("Delhi:",   np.std(delhi))
print("Kanpur:",  np.std(kanpur))
print("Mumbai:",  np.std(mumbai))
print("Noida:",   np.std(noida))
```

```python
print("\nCorrelation (Lucknow–Delhi):",  np.corrcoef(lucknow, delhi)[0,1])
print("Correlation (Lucknow–Kanpur):",  np.corrcoef(lucknow, kanpur)[0,1])
print("Correlation (Delhi–Kanpur):",    np.corrcoef(delhi, kanpur)[0,1])
print("Correlation (Delhi–Mumbai):",    np.corrcoef(delhi, mumbai)[0,1])
print("Correlation (Mumbai–Noida):",    np.corrcoef(mumbai, noida)[0,1])
```

```python
plt.figure()

plt.plot(months, lucknow, marker='o', label="Lucknow")
plt.plot(months, delhi,   marker='o', label="Delhi")
plt.plot(months, kanpur,  marker='o', label="Kanpur")
plt.plot(months, mumbai,  marker='o', label="Mumbai")
plt.plot(months, noida,   marker='o', label="Noida")

plt.title("Monthly Temperature Trend — All Cities")
plt.xlabel("Months")
plt.ylabel("Temperature (°C)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
print("Hottest month (Lucknow):", months[np.argmax(lucknow)])
print("Coldest month (Lucknow):", months[np.argmin(lucknow)])
print("Hottest month (Delhi):", months[np.argmax(delhi)])
print("Coldest month (Delhi):", months[np.argmin(delhi)])
print("Hottest month (Kanpur):", months[np.argmax(kanpur)])
print("Coldest month (Kanpur):", months[np.argmin(kanpur)])
print("Hottest month (Mumbai):", months[np.argmax(mumbai)])
print("Coldest month (Mumbai):", months[np.argmin(mumbai)])
print("Hottest month (Noida):", months[np.argmax(noida)])
print("Coldest month (Noida):", months[np.argmin(noida)])
```

```python
X = df[["Lucknow","Delhi","Kanpur","Mumbai","Noida"]].to_numpy()
monthly_avg = np.mean(X, axis=1)
months = df["Month"].astype(str).to_numpy()
print("\nMonthly Average Temp:")
for m, val in zip(months, monthly_avg):
    print(m, ":", round(val, 2))
```

```python
plt.figure()
plt.plot(months, monthly_avg, marker='o')
plt.title("Monthly Average Temperature (All Cities)")
plt.xlabel("Months")
plt.ylabel("Temperature (°C)")
plt.grid(True)
plt.show()
```

```python
print("Range (Lucknow):", np.max(lucknow) - np.min(lucknow))
```

```python
max_vals = [np.max(lucknow), np.max(delhi), np.max(kanpur), np.max(mumbai), np.max(noida)]
cities = ["Lucknow","Delhi","Kanpur","Mumbai","Noida"]
plt.figure()
plt.bar(cities, max_vals)
plt.title("Hottest Temperature Comparison")
plt.ylabel("Temp (°C)")
plt.show()
```

```python
city_cols = ['Lucknow','Delhi','Kanpur','Mumbai','Noida']
```

```python
pearson_corr = df[['Lucknow','Delhi','Kanpur','Mumbai','Noida']].corr()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(pearson_corr, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Pearson Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Delhi'], kde=True)
plt.title("Delhi temperature distribution")
plt.show()
```

```python
# Make sure Month is a column
if 'Month' not in df.columns:
    df = df.reset_index()     # converts index to column

# Now create df_long
df_long = df.melt(
    id_vars='Month',
    value_vars=city_cols,
    var_name='City',
    value_name='Temperature'
)
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=df_long, x='Month', y='Temperature', hue='City', marker='o')
plt.title("Monthly Temperature Trend - All Cities")
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.show()
```

```python
may_data = df[df['Month'] == 'May']
sns.barplot(x=city_cols, y=may_data[city_cols].values[0])
plt.title("Temperature Comparison in May")
plt.ylabel("Temperature (°C)")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df[city_cols])
plt.title("Temperature Distribution of All Cities")
plt.xlabel("City")
plt.ylabel("Temperature (°C)")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.violinplot(data=df[city_cols])
plt.title("City Temperature Distribution (Violin Plot)")
plt.xlabel("City")
plt.ylabel("Temperature (°C)")
plt.show()
```

```python
sns.scatterplot(x=df['Lucknow'], y=df['Delhi'])
plt.xlabel("Lucknow Temperature")
plt.ylabel("Delhi Temperature")
plt.title("Lucknow vs Delhi Temperature Scatter")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df_long, x='Month', y='Temperature', hue='City')
plt.title("Monthly Temperature Comparison - All Cities")
plt.show()
```

```python
sns.pairplot(df.drop(columns="Month"))
plt.suptitle("Relationship Between Temperature Variables", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df.drop(columns="Month").corr(), annot=True, cmap="viridis")
plt.title("Correlation Between Cities")
plt.show()
```

```python
seasons = {
    "Winter": ["Dec","Jan","Feb"],
    "Summer": ["Apr","May","Jun"],
    "Monsoon": ["Jul","Aug","Sep"]
}

season_avg = {}

for season, months in seasons.items():
    season_avg[season] = df[df["Month"].isin(months)].drop(columns="Month").mean().mean()

plt.figure(figsize=(7,7))
plt.pie(season_avg.values(), labels=season_avg.keys(), autopct='%1.1f%%')
plt.title("Seasonal Temperature Distribution")
plt.show()
```
